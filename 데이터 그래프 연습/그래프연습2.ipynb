```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
sns.set_style("whitegrid")
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
corr = tips.corr(method='pearson')
corr
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips, palette="Set3")
plt.show()
```

```python
plt.hist(tips['total_bill'], bins=10)
plt.title("Histogram of Total Bill")
plt.show()
```

```python
plt.scatter(tips['total_bill'], tips['tip'], s=tips['size']*10, c = tips['size'])
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.lmplot(x="total_bill", y="tip", data=tips, height=7)
plt.show()
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips, height=7)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="sex", y="tip", data=tips)
plt.show()
```

```python
sns.pairplot(tips, hue='sex')
plt.show()
```

```python
#    
flights = sns.load_dataset("flights")
flights.head()
```

```python
flights = flights.pivot("month", "year", "passengers")
flights.head()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(flights, annot=True, fmt="d")
plt.xlabel("year")
plt.ylabel("month")
plt.show()
```
